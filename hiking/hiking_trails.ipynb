```python
hiking_trails_swiss= []

for i in range(1,2):
    url = "https://www.wikiloc.com/trails/hiking/switzerland"
    scraper = cloudscraper.create_scraper()
    q1 = scraper.get(url)
    soup = BeautifulSoup(q1.text, "html.parser") 
    trails = soup.find_all('h2', class_='trail__title')
    hiking_trails_swiss.extend(trails) 
    soup.a.decompose()
    p_tags = soup.find_all('h2')
    for each in p_tags:
        print (each.text)
    # distance = soup.find_all('dd', class_='trail__detail__stats__value')
    # hiking_trails_swiss.extend(distance) 
len(hiking_trails_swiss)
```
